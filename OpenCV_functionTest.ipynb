#### matplotlib -> RGB RED GREEN BLUE (RGB)
#### OpenCV -> BLUE GREEN RED (BGR)
* 색상 체계 차이로 OpenCV로 읽은 이미지를 matplotlib으로 띄우면 색이 달라짐


```python
convert = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # from bgr to rgb
```

```python
plt.imshow(convert)
```

```python
gray = cv2.imread('D:/img/bird.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
gray.shape
```

```python
plt.imshow(gray)
```

```python
plt.imshow(gray)
```
